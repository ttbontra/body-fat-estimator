```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("bodyfat.csv")
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
def plotdistplots(col):
    plt.figure(figsize=(12,5))
    sns.distplot(df['BodyFat'],color='magenta',hist=False,label='BodyFat')
    sns.distplot(df[col],color='red',hist=False,label=col)
    plt.legend()
    plt.show()

cols = list(df.columns)
for i in cols:
    print(f'Distribution plots for {i} feature is shown below')
    plotdistplots(i)
    print("-"*100)
```

```python
import scipy.stats as stats
```

```python
def drawplots(df, cols):
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    plt.hist(df[cols], color='magenta')
    plt.subplot(1,2,2)
    stats.probplot(df[cols], dist='norm', plot=plt)
    
    plt.show()
cols = list(df.columns)
for i in range(len(cols)):
    print(f'Plots for {cols[i]} feature is shown below')
    drawplots(df, cols[i])
    print("-"*100)
```
