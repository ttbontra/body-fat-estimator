```python
def plotdistplots(col):
    plt.figure(figsize=(12,5))
    sns.distplot(df['BodyFat'],color='magenta',hist=False,label='BodyFat')
    sns.distplot(df[col],color='red',hist=False,label=col)
    plt.legend()
    plt.show()

cols = list(df.columns)
for i in cols:
    print(f'Distribution plots for {i} feature is shown below')
    plotdistplots(i)
    print("-"*100)
```

```python
import scipy.stats as stats
```

```python
def drawplots(df, cols):
    plt.figure(figsize=(12,5))
    plt.subplot(1,3,1)
    plt.hist(df[cols], color='magenta')
    plt.subplot(1,3,2)
    stats.probplot(df[cols], dist='norm', plot=plt)
    plt.subplot(1,3,3)
    sns.boxplot(y=df[cols], color='magenta')
    
    plt.show()
cols = list(df.columns)
for i in range(len(cols)):
    print(f'Plots for {cols[i]} feature is shown below')
    drawplots(df, cols[i])
    print("-"*100)
```

```python
upperlimit = []
lowerlimit = []
for i in df.columns:
    upperlimit.append(df[i].mean() + df[i].std()*4)
    lowerlimit.append(df[i].mean() - df[i].std()*4)
```

```python
cols = list(df.columns)
j = 0
for i in range(len(cols)):
    temp = df.loc[(df[cols[i]] > upperlimit[j])&(df[cols[i]] < lowerlimit[j])]
    j+=1
```

```python
def correlation(df, threshold):
    col_cor = set()
    cormat = df.corr()
    for i in range(len(cormat)):
        for j in range(i):
            if abs(cormat.iloc[i,j]) > threshold:
                colname = cormat.columns[i]
                col_cor.add(colname)
    return col_cor

ans = correlation(train, threshold= 0.8)
ans
```

```python
to_train = data[cols1]
to_train.head()
```

```python
from sklearn.model_selection import train_test_split,RandomizedSearchCV
from sklearn import metrics
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn import tree
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(to_train, test, test_size=0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
reg = DecisionTreeRegressor()
reg.fit(X_train, y_train)
plt.figure(figsize=(12,5))
tree.plot_tree(reg, filled=True)
```

```python
path = reg.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas
alphalist = []
for i in range(len(ccp_alphas)):
    reg = DecisionTreeRegressor(ccp_alpha=ccp_alphas[i])
    reg.fit(X_train, y_train)
    alphalist. append(reg)
```

```python
alphalist[1:5]
```

```python
trainscore = [alphalist[i].score(X_train, y_train) for i in range(len(alphalist))]
testscore = [alphalist[i].score(X_test, y_test) for i in range(len(alphalist))]
plt.xlabel('alpha')
plt.ylabel('accuracy')
plt.plot(ccp_alphas, trainscore, marker='o', color='red', label='train', drawstyle='steps-post')
plt.plot(ccp_alphas, testscore, marker='+', color='green', label='test', drawstyle='steps-post')
plt.legend()
plt.show()
```

```python
clf = DecisionTreeRegressor(ccp_alpha=1)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(f'Desicion Tree Normal Approach : {metrics.r2_score(y_test, y_pred)}')

rf = RandomForestRegressor(n_estimators=1000, ccp_alpha=1)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print(f'Random Forest Normal Approach : {metrics.r2_score(y_test, y_pred)}')
```
